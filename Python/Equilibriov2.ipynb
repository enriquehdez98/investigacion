```python
import pandas as pd

info = pd.read_csv("C:/Users/Dell/OneDrive - Universidad Autónoma del Estado de México/Documents/Maestria/Investigacion/MATLAB (Base de datos)/data OC/BDSinfo_OC.csv")
info['mat']=0
info['head']=0
info['info']=0
```

```python
import requests


for i in range(0,info["Trial"].size):
    print(i)
# URL del archivo que deseas descargar
    url = 'https://archive.physionet.org/atm/hbedb/' + list(info["Trial"].values)[i] + '/0/60/export/matlab/' + list(info["Trial"].values)[i] + 'm'
    url1= url + '.mat'
    url2 = url +'.hea'
    url3 = url + '.info'
    
    # Envía una solicitud GET para descargar el archivo
    response1 = requests.get(url1)

    # Verifica si la solicitud fue exitosa (código de estado 200)
    if response1.status_code == 200:
        # Abre un archivo local en modo binario para guardar el contenido
        with open('C:/Users/Dell/OneDrive - Universidad Autónoma del Estado de México/Documents/Maestria/Investigacion/MATLAB (Base de datos)/dataOC/' + list(info["Trial"].values)[i] + 'm.mat', 'wb') as file:
            file.write(response1.content)
        info.loc[info["Trial"]==list(info["Trial"].values)[i],"mat"] = 1
    else:
        print('Error al descargar el archivo. Código de estado:', response1.status_code)
        info.loc[info["Trial"]==list(info["Trial"].values)[i],"mat"] = 0

    # Envía una solicitud GET para descargar el archivo
    response2 = requests.get(url2)

    # Verifica si la solicitud fue exitosa (código de estado 200)
    if response2.status_code == 200:
        # Abre un archivo local en modo binario para guardar el contenido
        with open('C:/Users/Dell/OneDrive - Universidad Autónoma del Estado de México/Documents/Maestria/Investigacion/MATLAB (Base de datos)/dataOC/' + list(info["Trial"].values)[i] + 'm.hea', 'wb') as file:
            file.write(response2.content)
        info.loc[info["Trial"]==list(info["Trial"].values)[i],"hea"] = 1
    else:
        print('Error al descargar el archivo. Código de estado:', response2.status_code)
        info.loc[info["Trial"]==list(info["Trial"].values)[i],"hea"] = 0

    # Envía una solicitud GET para descargar el archivo
    response3 = requests.get(url3)

    # Verifica si la solicitud fue exitosa (código de estado 200)
    if response3.status_code == 200:
        # Abre un archivo local en modo binario para guardar el contenido
        with open('C:/Users/Dell/OneDrive - Universidad Autónoma del Estado de México/Documents/Maestria/Investigacion/MATLAB (Base de datos)/dataOC/' + list(info["Trial"].values)[i] + '.minfo', 'wb') as file:
            file.write(response3.content)
        info.loc[info["Trial"]==list(info["Trial"].values)[i],"info"] = 1
    else:
        print('Error al descargar el archivo. Código de estado:', response3.status_code)
        info.loc[info["Trial"]==list(info["Trial"].values)[i],"info"] = 0

        
```
